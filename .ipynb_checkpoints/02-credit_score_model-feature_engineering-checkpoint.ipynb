### Distribution Transformation

```python
from scipy.stats import yeojohnson


def log_transf (data, features):
    df_new = data.copy()
    for feature in features:
        yf, lam = yeojohnson(df[feature])
        df_new[feature] = yf

    return df_new
```

```python
test = log_transf(df_feat,num)
```
